a) nombre de la sede que acumulo la mayor cantidad de espectadores en eventos durante el certamen del 14 al 15 de cotubre inclusive.

```python
#me quedo con mes de octubre
#filtro = eventos [eventos['fecha'] >= 2018-10-01]
```

```python
#me quedo con dias 14 y 15
dias = eventos[(eventos['fecha'] == '2018-10-14') | (eventos['fecha'] == '2018-10-15')]
dias
```

```python
#hago un merge por id_locacion de dias con locaciones
sedes = dias.merge(locaciones, on = 'id_locacion', how = 'inner')
sedes
```

```python
#agrupo por sede quedandome con cantidad de espectadores y hago la suma por sede
sede = sedes.groupby('sede')['cantidad_espectadores'].sum().reset_index()
sede
```

```python
#ordeno y me quedo con el de mas espectadores
sede = sede.sort_values(by = 'cantidad_espectadores').tail(1)
sede
```

b) nombre del evento y nombre de la categoria deportiva de aquellos eventos cuya cantidad de espectadores supero la capacidad de la locacion, mas alla de la capacidad excedida.

```python
#me quedo con id_evento, cantidad_espectadores, nombre y id_categoria
even = eventos [["id_locacion" , "cantidad_espectadores", "nombre_evento", "id_categoria_deportiva"]]
even
```

```python
#creo una columna capacidad maxima
locaciones['cap_max'] = locaciones["capacidad"] + locaciones["capacidad_extendida"]
locaciones
```

```python
#me quedo con id_locacion y cap_max de locaciones
cap = locaciones[['id_locacion', 'cap_max']]
cap
```

```python
# uno #me quedo solo con id_categoria y nombre de categorias deportivas
cate = categorias_deportivas[['id_categoria_deportiva', 'nombre']]
cate
```

```python
#uno even con cap por id_locacion
union = cap.merge(even, how = 'inner', on = 'id_locacion')
union
```

```python
#creo una columna para ver que evento supero la capacidad maxima
union['capacidad_superada'] = (union['cantidad_espectadores']) > (union['cap_max'])
union
```

```python
#me quedo con los que la superaron
union = union[union['capacidad_superada'] == True]
union
```

```python
#me quedo nombre evento y id_categoria_deportiva
union = union[['id_categoria_deportiva', 'nombre_evento']]
union
```

```python
#lo uno con categorias deportivas por id
final = union.merge(categorias_deportivas, on='id_categoria_deportiva', how='inner')
final
```

```python
#renombro columna y me quedo solo con los nombres
final = final.rename(columns = {'nombre':'nombre_categoria_deportiva'})
final = final[["nombre_categoria_deportiva", "nombre_evento"]]
final
```
